### Exploratory Data Analysis (EDA)
Performing EDA on the dataset to understand its structure, distribution, and potential issues.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
csv_file_path = r"abuzar_data.csv"
data = pd.read_csv(csv_file_path)

# Display the first few rows
data.head()
```

```python
# Display dataset information
data.info()
```

```python
# Check for missing values
missing_values = data.isnull().sum()
print("Missing Values:")
print(missing_values[missing_values > 0])
```

```python
# Display basic statistics
data.describe()
```

```python
# Analyze class distribution for 'fight_result' column
print("Class Distribution:")
print(data['fight_result'].value_counts())
```

```python
# Plot correlation matrix (numeric columns only)
numeric_data = data.select_dtypes(include=['number'])
plt.figure(figsize=(12, 8))
sns.heatmap(numeric_data.corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Matrix")
plt.show()
```

```python
# Plot distribution of key features
features = ['health', 'x_coord', 'y_coord', 'timer']
for feature in features:
    plt.figure(figsize=(8, 4))
    sns.histplot(data[feature], kde=True, bins=30)
    plt.title(f"Distribution of {feature}")
    plt.show()
```

```python
# Analyze categorical features
categorical_features = ['is_jumping', 'is_crouching', 'is_player_in_move']
for feature in categorical_features:
    plt.figure(figsize=(6, 4))
    sns.countplot(x=data[feature])
    plt.title(f"Count of {feature}")
    plt.show()
```
